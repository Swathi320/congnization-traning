```python
def add (a,b,c=0):
    return a+b+c
print(add(10,20))
print(add(10,20,30))
```

```python
def multiply ( a,b,c=0): 
    return a*b*c
print(multiply(10,20))
print(multiply(10,20,30))
```

```python
def add(*numbers):
    print(sum(numbers))
add(10,20)
add(10,20,30)
add(10,20,30,40)
```
