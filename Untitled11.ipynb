```python
import pandas as pd
student={
       "Name":["swathi","sravani","akshaya","sreeja"],
        "Age":[20,20,19,25]
}
result=pd.DataFrame(student)
print(result)
print(result.duplicated())
print(result.drop_duplicates())
```

```python
Employee={
    "Name":["Mahesh","Ritik","Ankur"],
    "salary":[50000,60000,85000]
}
result=pd.DataFrame(Employee)
print(result)
print(result[result["salary"]>50000])
print(result.query("salary>50000"))
print(result.query("salary>50000 and salary<85000"))
```
