```python
import matplotlib.pyplot as plt
import numpy as np
products=["Mobile","Laptop","TV"]
sales_2024=[2000,300,800]
sales_2025=[4000,350,678]
sales_2026=[5400,500,879]
x=np.arange(len(products))
width=0.35
plt.bar(x-width/3,sales_2024,width,label="2024")
plt.bar(x+width/3,sales_2025,width,label="2025")
plt.bar(x*width/2,sales_2026,width,label="2026")
plt.xlabel("products")
plt.xlabel("sales")
plt.title("sales comparison: 2024 vs 2025")
plt.legend()
plt.show()

          
```

```python
import matplotlib.pyplot as plt
import numpy as np
products=["mobile","laptops","TV"]
online=[2000,400,4000]
offline=[3000,340,3600]
wholesale=[15000,300,3300]
plt.bar(products,online,label="online")
plt.bar(products,offline,label="offline")
bottom_value=[online[i]+offline[i] for i in range(len(products))]
plt.bar(products,wholesale,bottom=bottom_value,label="wholesale")
plt.xlabel("products")
plt.ylabel("Revenue")
plt.title("Revenue of the Distrubuted sales")
plt.legend()
plt.xticks(x,products)
plt.show()
```
