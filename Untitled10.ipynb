```python
import numpy as np
arr=np.array([10,20,30,40,50])
print(arr)
print(max(arr))
print(min(arr))
print(sum(arr))
print(arr[2])
print(np.vstack((arr)))
print(np.hstack((arr)))
print(arr+5)
print(arr[::2])
print(arr[10:50:20])
print(np.power(arr,2))
print(np.square(arr*arr))
x=np.sort(arr)
print(x)
y=np.where(arr==30)
print(y)
for i in range(1,5):
    print(arr[i])
for i in np.nditer(arr):
    print(i)
c=np.concatenate([arr])
print(c)
```
