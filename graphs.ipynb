```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
months=["jan","feb","mar","apr","may"]
sales=[100,200,150,400,300]
sns.lineplot(x=months,y=sales,markers=0)
plt.xlabel(months)
plt.ylabel(sales)
plt.title('monthly sales trade')
plt.show()
```

```python
import plotly.express as px
import pandas as pd
df=pd.DataFrame({
    'months':['jan','feb','mar','apr','may'],
    'sales':[100,200,150,400,300]})
fig=px.line(df,x='months',y='sales',markers=True)
fig.show()
    
```
