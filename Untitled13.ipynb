```python
import numpy as np
products=["Mobile","Laptop","Tablet"]
sales_2024=[500,300,200]
sales_2025=[800,400,200]
x=np.arange(len(products))
width=0.35
plt.bar(x-width/2,sales_2024,width)
plt.bar(x-width/2,sales_2025,width)
plt.show()
```

```python
products=["Mobile","Laptop","Tablet","TV"]
online=[200,300,400,500]
offline=[100,500,400,300]
plt.bar(products,online)
plt.bar(products,offline,bottom=online)
plt.show()
```

```python
import random 
developers=list(range(1,51))
testers=list(range(51,81))
mangers=list(range(81,100))
sample_dev=random.sample(developers,5)
sample_test=random.sample(testers,3)
sample_mrg=random.sample(mangers,2)
final_sample=sample_dev+sample_test+sample_mrg
print(final_sample)
```

```python
clusters={
        "hyderabad":[1,2,3,4,5],
         "pune":[6,7,8,9,10],
         "chennai":[11,12,13,14,15]
}
select_clusters=["hyderabad","chennai"]
sample=[]
for cluster in select_clusters:
    sample.extend(clusters[cluster])
    print("cluster sample:",sample)
```

```python
networks={
          "A":["B","C"],
          "B":["D"],
          "C":["E"]
}
sample=["A"]
sample.extend(networks["A"])
sample.extend(networks["B"])
print("snowBall sampling:",sample)
```
